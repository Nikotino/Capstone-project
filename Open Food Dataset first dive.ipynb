<h4> Load and preprocess the Open Food Facts dataset </h4> 

```python
data=pd.read_csv('en.openfoodfacts.org.products.tsv', sep="\t")
```

```python
data.head()
```

```python
print(data.loc[3].to_string())
```

```python
data_short=data[['product_name', 'serving_size', 'energy_100g', 'fat_100g', 
                 'saturated-fat_100g', 'carbohydrates_100g','sugars_100g', 'proteins_100g', 'salt_100g']]
```

```python
data_short=data_short.dropna()
```

```python
def cut(a):
    a=a[:a.find(' ')]
    a=a.replace(",", '.')
    try: 
        return np.float(a)
    except: return 1
    
```

```python
data_short['serving_size']=pd.Series([cut(a) for a in data_short.serving_size])
```

```python
data_short.head()
```

```python
for column in ['fat_100g', 'saturated-fat_100g', 'carbohydrates_100g', 'sugars_100g', 'proteins_100g']:
    data_short=data_short[(data_short[column]>=0)&(data_short[column]<=100)]
```

```python
for column in ['energy_100g', 'fat_100g', 'saturated-fat_100g', 'carbohydrates_100g', 'sugars_100g', 'proteins_100g', 'salt_100g']:
    data_short[column]=data_short[column]*data_short['serving_size']/100
```

```python
data_short.head()
```

```python
data_short=data_short.dropna()
```

```python

data_short=data_short.reset_index()
data_short=data_short.drop('index', axis=1)
```
